```python
import matplotlib.pyplot as plt
import pandas as pd
count = 0
mean = {}
sd = {}
o = [1,2,3,4,5,6,7,"7b",8,"8b"]

for i in o:
   file_name = "./final/agent" + str(i)+'.csv'
   df = pd.read_csv(file_name)
   mean[i] =  df["success"][0]
   sd[i] =  df["success"][1]

print(mean)
print(sd)
print() 
k = [i for i in range(len(mean.keys()))]
for i, j in zip(k,list(mean.values()) ):
   plt.text(i + 0.050, j + 0.0010, "({}, {})".format(list(mean.keys())[i], round(j, 2)))
plt.xlabel('Agent Numbers')
plt.ylabel('Average Success Rate')
plt.title("COMPARE ALL AGENTS")
# plt.xticks()
plt.plot(list(mean.keys()), list(mean.values()))
plt.errorbar(k, list(mean.values()), list(sd.values()), color = 'red',marker='s',ecolor = 'b')
plt.gcf().set_size_inches(11.2775330396, 7.04845814978) # The MacbookPro 13.3 inches size
plt.savefig("./graphs/compare_all.png", dpi=227) # The MacbookPro 13.3 inches dpi
plt.show()
```
